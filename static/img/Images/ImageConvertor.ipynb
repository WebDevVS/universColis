```python
import shutil
import os

def filtered_copytree(src, dst):
    def ignore_gdocs(dir, files):
        ignored = []
        for f in files:
            if f.endswith('.gdoc') or f.endswith('.gsheet') or f.endswith('.gslides'):
                ignored.append(f)
        return ignored

    shutil.copytree(src, dst, ignore=ignore_gdocs)

source = "/content/drive/MyDrive/Colab Notebooks/WEBSITE /Images"
backup = "/content/drive/MyDrive/Colab Notebooks/WEBSITE /Images_backup"

if not os.path.exists(backup):
    filtered_copytree(source, backup)
    print("✅ Backup created at:", backup)
else:
    print("⚠️ Backup already exists. Skipping copy.")
```

```python
import os
from PIL import Image

converted_count = 0  # Global counter

def get_file_size_kb(filepath):
    return round(os.path.getsize(filepath) / 1024, 2)

def optimize_image(input_path, output_path, max_width=1200, quality=60):
    global converted_count
    try:
        img = Image.open(input_path)
        original_format = img.format

        original_size = get_file_size_kb(input_path)

        # Resize if needed
        if img.width > max_width:
            ratio = max_width / img.width
            new_height = int(img.height * ratio)
            img = img.resize((max_width, new_height), Image.LANCZOS)

        img = img.convert("RGB")  # Strip alpha/metadata
        img.save(output_path, "WEBP", quality=quality, method=6)

        new_size = get_file_size_kb(output_path)

        converted_count += 1
        print(f"✅ {input_path}")
        print(f"   📸 Format: {original_format} | 🧱 Original: {original_size} KB ➜ WebP: {new_size} KB\n")

    except Exception as e:
        print(f"❌ Failed: {input_path} - {e}")

def convert_all_images(root_folder):
    supported_exts = (".jpg", ".jpeg", ".png", ".webp")
    global converted_count
    converted_count = 0

    for dirpath, _, filenames in os.walk(root_folder):
        if "Images_backup" in dirpath:
            continue  # skip backup folders

        for filename in filenames:
            if filename.lower().endswith(supported_exts):
                input_path = os.path.join(dirpath, filename)
                base_name = os.path.splitext(filename)[0]
                output_path = os.path.join(dirpath, base_name + "_webp.webp")

                if os.path.exists(output_path):
                    continue

                optimize_image(input_path, output_path)

    print(f"\n🎯 Total images converted: {converted_count}")
```

```python
your_drive_path = "/content/drive/MyDrive/Colab Notebooks/WEBSITE /Images"
convert_all_images(your_drive_path)
```

```python
# Celean the orefix , doubles , old images and keep the non  images files and the new created compressed images


import os

def delete_only_old_images(root_folder):
    deleted = 0
    for dirpath, _, filenames in os.walk(root_folder):
        if "Images_backup" in dirpath:
            continue  # skip backup

        for filename in filenames:
            lower = filename.lower()

            # Only target image types
            if lower.endswith((".jpg", ".jpeg", ".png", ".webp")) and not lower.endswith("_webp.webp"):
                base_name = os.path.splitext(filename)[0]
                webp_version = base_name + "_webp.webp"

                if webp_version in filenames:
                    file_path = os.path.join(dirpath, filename)
                    try:
                        os.remove(file_path)
                        deleted += 1
                        print(f"🗑️ Deleted original image: {file_path}")
                    except Exception as e:
                        print(f"❌ Error deleting {file_path}: {e}")

    print(f"\n✅ Finished. Deleted {deleted} original image files.")

# Your folder path
your_drive_path = "/content/drive/MyDrive/Colab Notebooks/WEBSITE /Images"
delete_only_old_images(your_drive_path)
```

```python
# fixing _webp.webp into .webp

import os

def rename_webp_files_clean(root_folder):
    renamed = 0
    for dirpath, _, filenames in os.walk(root_folder):
        if "Images_backup" in dirpath:
            continue  # Never touch backup

        for filename in filenames:
            if filename.lower().endswith("_webp.webp"):
                old_path = os.path.join(dirpath, filename)
                new_name = filename.replace("_webp.webp", ".webp")
                new_path = os.path.join(dirpath, new_name)

                # Safety check — skip if a file with new name already exists
                if not os.path.exists(new_path):
                    os.rename(old_path, new_path)
                    renamed += 1
                    print(f"✏️ Renamed: {filename} ➜ {new_name}")
                else:
                    print(f"⚠️ Skipped (already exists): {new_name}")

    print(f"\n✅ Finished. Renamed {renamed} files to clean .webp names.")

# Run the rename
your_drive_path = "/content/drive/MyDrive/Colab Notebooks/WEBSITE /Images"
rename_webp_files_clean(your_drive_path)


```
